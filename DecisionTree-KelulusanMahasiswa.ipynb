```python
#Import library numpy, pandas dan scikit-learn
import numpy as np
import pandas as pd
from sklearn import tree
```

```python
#Membaca Dataset dari File ke Pandas dataFrame
irisDataset = pd.read_csv('kelulusanmhs.csv', delimiter=';', header=0)
```

```python
irisDataset.head()
```

```python
#Mengubah kelas (kolom "Species") dari String ke Unique-Integer
irisDataset["KELULUSAN"] = pd.factorize(irisDataset.KELULUSAN)[0]
```

```python
irisDataset.head()
```

```python
print(irisDataset)
```

```python
#Menghapus kolom "Id"
irisDataset = irisDataset.drop(labels="ID", axis=1)
```

```python
print(irisDataset)
```

```python
#Mengubah dataFrame ke array Numpy
irisDataset = irisDataset.to_numpy()
```

```python
print(irisDataset)
```

```python
#Membagi Dataset => 80 baris data untuk training dan 20 baris data untuk testing
dataTraining = np.concatenate((irisDataset[0:40, :], irisDataset[50:90, :]),
                              axis=0)
dataTesting = np.concatenate((irisDataset[40:50, :], irisDataset[90:100, :]),
                             axis=0)
```

```python
print(dataTesting)
len(dataTesting)
```

```python
#Memecah Dataset ke Input dan Label
inputTraining = dataTraining[:, 0:9]
inputTesting = dataTesting[:, 0:9]
labelTraining = dataTraining[:, 9]
labelTesting = dataTesting[:, 9]
```

```python
#Mendefinisikan Decision Tree Classifier
model = tree.DecisionTreeClassifier()
```

```python
#Mentraining Model
from sklearn.impute import SimpleImputer

# Membuat objek SimpleImputer dengan strategi pengisian 'mean'
imputer = SimpleImputer(strategy='mean')

# Menggunakan SimpleImputer untuk mengisi nilai-nilai yang hilang dalam data
inputTraining = imputer.fit_transform(inputTraining)
model = model.fit(inputTraining, labelTraining)
```

```python
#Memprediksi Input Data Testing
hasilPrediksi = model.predict(inputTesting)
print("Label Sebenarnya : ", labelTesting)
print("Hasil Prediksi : ", hasilPrediksi)
```

```python
#Menghitung Akurasi
prediksiBenar = (hasilPrediksi == labelTesting).sum()
prediksiSalah = (hasilPrediksi != labelTesting).sum()
print("Prediksi Benar :", prediksiBenar, "data")
print("Prediksi Salah :", prediksiSalah, "data")
print("Akurasi :", prediksiBenar/(prediksiBenar+prediksiSalah) * 100, "%")
```
